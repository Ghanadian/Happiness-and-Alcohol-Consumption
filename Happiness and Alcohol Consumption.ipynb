## Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

## Import the dataset

```python
url='HappinessAlcoholConsumption.csv'
df=pd.read_csv(url)
df.head()
```

## All data are for 2016 add this after the name of the country.

```python
def add_year(text):
    return str(text) +'_2016'
df['Country']=df['Country'].map(add_year)
df
```

## Create a lambda function that will capitalize strings.

```python
capitalizer = lambda x: x.capitalize()
```

## capitalize Hemisphere.

```python
df['Hemisphere']= df['Hemisphere'].apply(capitalizer)
df
```

## Print the last elements of the data set.

```python
df.tail()
```

## In happy countries, is the mean Spirit_PerCapita  more or less?( Assuming that a happy country is a country where the HappinessScore is greater than 5)

```python
#Create a function called majority_happy that returns a boolean value to a new column called Happiness 
#(Consider majority_happy as more than 5)

def majority_happy(x):
    if x>5:
        return True
    else:
        return False
            
df['Happiness']=df['HappinessScore'].apply(majority_happy)
df
```

```python
df.groupby('Happiness').agg({'Spirit_PerCapita': 'mean'})
```
